## Prototype: voorspellen of de patiënt zwanger is met gestructureerde, ongestructureerde en semigestructureerde datasetsoorten

**Achtergrond**

Uit de deelvragen die in het onderzoeksverslag zijn uitgewerkt blijk dat semigestructureerde data het meeste inzicht biedt in de medische sector. Daarnaast kan semigestructureerde data, combinatie van gestructureerde en ongestructureerde data, interpreteerbaar zijn voor mensen met kennis uit de medische sector. Dit inzicht maakt dat het nuttig is om te kijken of machine learning modellen die zowel gestructureerde, semigestructureerde als ongestructureerde data verwerken nauwkeurig en interpreteerbaar zijn. 

**Methode**

Het prototype waarin voorspeld wordt of een patiënt zwanger is, is opgezet om te onderzoeken of de datasetsoort invloed heeft op de mate van interpreteerbaarheid. Door de drie datasetsoorten in de AutoML H2O te trainen wordt bepaald welk model het best presteert. Er is gekozen voor de AutoML H2O omdat uit het onderzoek van Rik Smink blijkt dat deze tool gratis en makkelijk in gebruik is. Het best presterende model wordt vervolgens getraind. Met het berekenen van de nauwkeurigheid en de F1-score wordt bepaald of het model nauwkeurig is. 

De data is verkregen van een openbare gezichten dataset van Kaggle (Male and female faces dataset, 2020). Hieruit is een selectie gemaakt van 30 vrouwen en 19 mannen. Op basis van deze data is de gestructureerde data, bestaand uit leeftijd en geslacht beschreven. Hierin is de leeftijd geschat en het geslacht is bepaald aan de hand van de afbeeldingen. Voor de ongestructureerde data is de breedte, lengte, grootte en pixeldichtheid van de afbeeldingen beschreven door de informatie van de afbeeldingen te bekijken. De voorspelling of de patiënt zwanger is, is bedacht en beschreven in iedere datasetsoort. 



#### Gestructureerde data
Als eerst wordt de gestructureerde dataset in de AutoML getraind die bekijkt welk machine learning model het best presteerd. Het best presterende machine learning model wordt in het bestand GBMmodellen.ipynb getraind, zodat er bekeken kan worden of de resultaten interpreteerbaar zijn.

```python
#importeren van de benodigde libraries
import h2o
from h2o.automl import H2OAutoML
import pandas as pd
```

```python
#starten van H2O instantie
h2o.init()
```

```python
#data importeren
gezichtendata = h2o.import_file('gezichtendata.csv', sep = ';')
features = ['geslacht', 'leeftijd']
output = 'zwanger'
```

```python
#de data opsplitsen in een test en trainings data
train, test = gezichtendata.split_frame(ratios=[0.8])
```

```python
#het model bouwen
aml = H2OAutoML(max_models = 15, max_runtime_secs=60, seed = 1234)
aml.train(x = features, y = output, training_frame = train)
```

```python
#leaderbord bekijken van best presterende modellen
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
preds = aml.predict(test)
preds
```

Uit de output van de AutoML komen verschillende resultaten naar voren. Zo speelt de leeftijd van de persoon een iets grotere rol (52,4%) in de voorspelling of de persoon zwanger is dan het geslacht (47,6%). 

Daarnaast wordt bekeken welk model het beste presteerd. Zeven machine learning modellen presteren beter dan de kans dat iemand zwanger is in de dataset (8/49 = 0,1633). Van de zeven machine learning modellen presteerd het Gradient Boosting Model (GBM) het best met een voorspellings waarde van 0,4205. Dit maakt dat het Gradient Boosting model het best presteerd voor de gestructureerde dataset. Het model wordt in het bestand GBMmodellen.ipynb getraind, zodat er bekeken kan worden of de resultaten interpreteerbaar zijn.

#### Ongestructureerde data
Vervolgens wordt de ongestructureerde data in de AutoML getraind die bekijkt welk machine learning model het best presteerd. Voordat de data in een AutoML getriand kan worden moeten de afbeeldingen eerst worden omgezet naar een CSV bestand. Dit is gedaan door verschillende kenmerken van de afbeelding mee te nemen, zoals breedte, lengte, grootte van het bestand en pixeldichtheid. 

```python
#starten van H2O instantie
h2o.init()
```

```python
#data importeren
afbeeldingdata = h2o.import_file('afbeeldingdata.csv', sep = ';')
features = ['breedte', 'lengte', 'grootte', 'pixeldichtheid']
output = 'zwanger'
```

```python
#de data opsplitsen in een test en trainings data
train, test = afbeeldingdata.split_frame(ratios=[0.8])
```

```python
#het model bouwen
aml = H2OAutoML(max_models = 15, max_runtime_secs=60, seed = 1234)
aml.train(x = features, y = output, training_frame = train)
```

```python
#leaderbord bekijken van best presterende modellen
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
preds = aml.predict(test)
preds
```

De resultaten van de AutoML laten zien dat de pixeldichtheid de belangrijkste variabele is in de voorspelling of iemand zwanger is met een percentage van 55,13%. Daarnaast zijn er 4 modellen die beter presteren dan de kans van 0,1633 dat iemand zwanger is. Van de vier modellen presteert het Gradient Boosting Model (GBM) het best met een voorspellingswaarde van 0,2097. Dit maakt dat het Gradient Boosting model het best presteert voor de ongestructureerde dataset. Het model wordt in het bestand GBMmodellen.ipynb getraind, zodat er bekeken kan worden of de resultaten interpreteerbaar zijn.  

#### Semigestructureerde data
Door de gestructureerde en ongestructureerde data samen te voegen en in een AutoML te trainen kan er bekeken worden welk machine learning model het best presteerd. 

```python
#starten van H2O instantie
h2o.init()
```

```python
#data importeren
gezichten_afbeelding = h2o.import_file('gezichten_afbeelding.csv', sep = ';')
features = ['geslacht', 'leeftijd', 'breedte', 'lengte', 'grootte', 'pixeldichtheid']
output = 'zwanger'
```

```python
#de data opsplitsen in een test en trainings data
train, test = gezichten_afbeelding.split_frame(ratios=[0.8])
```

```python
#het model bouwen
aml = H2OAutoML(max_models = 15, max_runtime_secs=60, seed = 1234)
aml.train(x = features, y = output, training_frame = train)
```

```python
#leaderbord bekijken van best presterende modellen
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
preds = aml.predict(test)
preds
```

Opvallend aan de resultaten van het trainen van semigestructureerde data in de AutoML is dat de variabele pixeldichtheid de grootste rol speelt (36,5%). De variabelen leeftijd (20,7%) en geslacht (15,7%) spelen een minder belangrijke rol. De verwachting was dat leeftijd en geslacht een grotere invloed zouden hebben op de voorspelling in de semigestructureerde dataset. De rede dat de pixeldichtheid meer invloed heeft op de voorspelling kan komen doordat de waarden vaak hetzelfde zijn. Hierdoor kan het H2O model beïnvloed zijn. 

Daarnaast zijn er 2 modellen die beter presteren dan de kans van 0,1633 dat iemand zwanger is. Het Gradient Boosting Model heeft een voorspellingswaarde van 0,4328 en is daarmee het best presterende model voor de semigestructureerde data. Het model wordt in het bestand GBMmodellen.ipynb getraind, zodat er bekeken kan worden of de resultaten interpreteerbaar zijn.
